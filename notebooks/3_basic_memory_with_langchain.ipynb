```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
```

```python
prompts = [
    {
    "role":"user", "content":"hello, my name is Shubham kumar."
},
    {
    "role":"ai", "content":"hello, Shubham kumar! how can i assist you today."
},
    {
    "role":"user", "content":"What is My name ?"
},
]

res = llm.invoke(prompts)
```

```python
res.content
```

### Manage History

```python
history = []
while True:
    query = input("User: ")
    if query.lower() in ["exit","quit","bye"]:
        print("Good Bye")
        break

    history.append({"role":"user", "content":query})
    print("User: ", query)

    res = llm.invoke(history)
    history.append({"role":"ai", "content":res.content})
    print("AI: ", res.content, "\n")
```
